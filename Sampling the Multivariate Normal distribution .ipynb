```python
X = np.random.randn(1000, 2)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
mu =  np.array([2,3])
cov = np.array([[2, 0.5], [0.5,1]])
```

```python
L = np.linalg.cholesky(cov)
```

```python
L
```

```python
Y = mu + X @ L.T
```

```python
plt.scatter(Y[:,0], Y[:,1])
```
